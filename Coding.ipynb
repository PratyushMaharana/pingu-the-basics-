```python
def is_palindrome():
    # Input the string from the user
    input_string = input("Enter a string: ")

    # Initialize pointers for comparison
    left = 0
    right = len(input_string) - 1

    # Check for palindrome by comparing characters from both ends
    while left < right:
        if input_string[left] != input_string[right]:
            print("The string is NOT a palindrome.")
            return  # Exit if mismatch is found
        left += 1
        right -= 1

    # If the loop completes, the string is a palindrome
    print("The string IS a palindrome.")

# Call the function
is_palindrome()
```

```python
def capitalize_every_other_letter():
    # Input the string from the user
    input_string = input("Enter a string: ")

    # Initialize an empty result string
    result = ""

    # Loop through the characters of the string
    for i, char in enumerate(input_string):
        if i % 2 == 0:  # For every other letter (even index)
            result += char.upper()  # Convert to uppercase
        else:  # For odd index
            result += char.lower()  # Convert to lowercase (to ensure consistent format)

    # Display the result
    print("Modified String:", result)

# Call the function
capitalize_every_other_letter()
```

```python
# Existing dictionary with roll numbers as keys and marks as values
students = {
    101: 85,
    102: 90,
    103: 78
}

# Function to add new student data to the dictionary
def add_student(roll_no, marks):
    if roll_no in students:
        print(f"Roll number {roll_no} already exists with marks {students[roll_no]}.")
    else:
        students[roll_no] = marks
        print(f"Added roll number {roll_no} with marks {marks}.")

# Adding new students
while True:
    print("\n1. Add a student")
    print("2. View all students")
    print("3. Exit")
    choice = input("Enter your choice: ")

    if choice == "1":
        try:
            roll_no = int(input("Enter the roll number: "))
            marks = int(input("Enter the marks: "))
            add_student(roll_no, marks)
        except ValueError:
            print("Invalid input. Please enter valid numbers.")
    elif choice == "2":
        print("\nStudents data:")
        for roll_no, marks in students.items():
            print(f"Roll No: {roll_no}, Marks: {marks}")
    elif choice == "3":
        print("Exiting the program.")
        break
    else:
        print("Invalid choice. Please try again.")
```

```python
# Simple Menu-Driven Program without using built-in functions

def display_menu():
    print("Menu:")
    print("1. Addition")
    print("2. Subtraction")
    print("3. Multiplication")
    print("4. Division")
    print("5. Exit")

def get_user_choice():
    choice = input("Enter your choice (1-5): ")
    return choice

def get_numbers():
    num1 = input("Enter the first number: ")
    num2 = input("Enter the second number: ")
    return num1, num2

def add(num1, num2):
    result = num1 + num2
    return result

def subtract(num1, num2):
    result = num1 - num2
    return result

def multiply(num1, num2):
    result = num1 * num2
    return result

def divide(num1, num2):
    if num2 != 0:
        result = num1 / num2
        return result
    else:
        return "Error: Division by zero is not allowed."

def main():
    while True:
        display_menu()
        choice = get_user_choice()

        if choice == '5':
            print("Exiting the program. Goodbye!")
            break

        if choice in ['1', '2', '3', '4']:
            num1, num2 = get_numbers()
            num1 = float(num1)
            num2 = float(num2)

            if choice == '1':
                result = add(num1, num2)
                print(f"The result of addition is: {result}")
            elif choice == '2':
                result = subtract(num1, num2)
                print(f"The result of subtraction is: {result}")
            elif choice == '3':
                result = multiply(num1, num2)
                print(f"The result of multiplication is: {result}")
            elif choice == '4':
                result = divide(num1, num2)
                print(f"The result of division is: {result}")
        else:
            print("Invalid choice. Please select a valid option (1-5).")

if __name__ == "__main__":
    main()
```

```python
def display_list(lst):
    print("Current List:", lst)

def insert_number(lst):
    num = int(input("Enter the number to insert: "))
    lst.append(num)
    print(f"Number {num} inserted.")

def delete_number(lst):
    if not lst:
        print("List is empty. Nothing to delete.")
        return
    num = int(input("Enter the number to delete: "))
    if num in lst:
        lst.remove(num)
        print(f"Number {num} deleted.")
    else:
        print(f"Number {num} not found in the list.")

def insert_at_position(lst):
    num = int(input("Enter the number to insert: "))
    pos = int(input("Enter the position to insert at: "))
    if 0 <= pos <= len(lst):
        lst.insert(pos, num)
        print(f"Number {num} inserted at position {pos}.")
    else:
        print("Invalid position.")

def delete_from_position(lst):
    if not lst:
        print("List is empty. Nothing to delete.")
        return
    pos = int(input("Enter the position to delete from: "))
    if 0 <= pos < len(lst):
        deleted_num = lst.pop(pos)
        print(f"Number {deleted_num} deleted from position {pos}.")
    else:
        print("Invalid position.")

def main():
    lst = []
    while True:
        print("\nMenu:")
        print("1. Display List")
        print("2. Insert Number")
        print("3. Delete Number")
        print("4. Insert at Specific Position")
        print("5. Delete from Specific Position")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == '1':
            display_list(lst)
        elif choice == '2':
            insert_number(lst)
        elif choice == '3':
            delete_number(lst)
        elif choice == '4':
            insert_at_position(lst)
        elif choice == '5':
            delete_from_position(lst)
        elif choice == '6':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")

if __name__ == "__main__":
    main()
```

```python
# Taking input as a space-separated tuple of numbers
tpl = tuple(map(int, input("Enter the tuple elements separated by space: ").split()))

# Dictionary to store element frequencies
freq = {}

# Loop to count occurrences of each element
for num in tpl:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

# Creating tuples for unique and duplicate elements
unique_elements = tuple(k for k, v in freq.items() if v == 1)
duplicate_elements = tuple(k for k, v in freq.items() if v > 1)

# Printing results
print("Frequencies of elements:", freq)
print("Tuple of unique elements:", unique_elements)
print("Tuple of duplicate elements:", duplicate_elements)
```

```python
def reverse_string(input_str):
    reversed_str = ""  # Initialize an empty string to store the reversed result
    length = 0

    # Calculate the length of the string manually
    for _ in input_str:
        length += 1

    # Iterate through the string in reverse order
    for i in range(length - 1, -1, -1):
        reversed_str += input_str[i]  # Append characters to the reversed string

    return reversed_str

# Example usage
input_str = "hello"
result = reverse_string(input_str)
print("Original String:", input_str)
print("Reversed String:", result)
```

```python
def is_palindrome(input_str):
    length = 0

    # Calculate the length of the string manually
    for _ in input_str:
        length += 1

    # Compare characters from the start and end
    for i in range(length // 2):  # Only need to check half the string
        if input_str[i] != input_str[length - 1 - i]:
            return False  # If characters don't match, it's not a palindrome

    return True  # If all characters match, it's a palindrome

# Example usage
input_str = "lolololol"
result = is_palindrome(input_str)
print(f"Is '{input_str}' a palindrome? {result}")
```

```python
def count_occurrences(line, substring):
    count, i = 0, 0
    while i <= len(line) - len(substring):
        if line[i:i+len(substring)] == substring:
            count += 1
        i += 1
    return count

line = input("Enter the line: ")
substring = input("Enter the substring: ")
print(f"The substring '{substring}' occurs {count_occurrences(line, substring)} time(s) in the line.")
```

```python
def dict_operations():
    my_dict = {'a': 1, 'b': 2, 'c': 3}
    print("Dictionary:", my_dict)
    print("1. Access value\n2. Add key-value\n3. Delete key")
    choice = int(input("Enter choice: "))
    if choice == 1:
        key = input("Enter key: ")
        print("Value:", my_dict.get(key, "Key not found"))
    elif choice == 2:
        key, value = input("Enter key: "), int(input("Enter value: "))
        my_dict[key] = value
        print("Updated dictionary:", my_dict)
    elif choice == 3:
        key = input("Enter key to delete: ")
        my_dict.pop(key, None)
        print("Updated dictionary:", my_dict)

def main():
    while True:
        print("\n1. Dictionary Operations\n2. Exit")
        choice = int(input("Enter choice: "))
        if choice == 1:
            dict_operations()
        elif choice == 2:
            break
        else:
            print("Invalid choice!")

if __name__ == "__main__":
    main()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Define x values
x = np.linspace(-1, 7, 400)

# Equations of the lines
y1 = (12 - 2*x) / 3  # From 2x + 3y = 12
y2 = x - 1           # From x - y = 1

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the lines
plt.plot(x, y1, label=r'$2x + 3y = 12$', color='blue')
plt.plot(x, y2, label=r'$x - y = 1$', color='green')

# Shade the feasible region
y1_shade = np.minimum(y1, 4)  # Bound y1 shading to y >= 0
plt.fill_between(x, 0, y1_shade, where=(x >= 0) & (y1 >= 0), color='blue', alpha=0.3)
plt.fill_between(x, y2, 10, where=(x >= 0) & (y2 >= 0), color='green', alpha=0.3)

# Highlight points of interest
plt.scatter([0, 6, 1], [4, 0, 0], color='red', zorder=3, label="Intercepts & Intersection")

# Labels and settings
plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
plt.xlim(-1, 7)
plt.ylim(-2, 6)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.title("Graphical Solution of the Inequalities")
plt.grid()

# Show the plot
plt.show()
```
